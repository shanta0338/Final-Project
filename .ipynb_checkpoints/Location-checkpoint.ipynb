```python
# ============================================================
# Predict location_name and country — Multi-Class Classification
# ============================================================

import mlflow
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from torch.utils.data import TensorDataset, DataLoader

# ── Device ────────────────────────────────────────────────────
if torch.cuda.is_available():
    device = torch.device("cuda")
elif torch.backends.mps.is_available():
    device = torch.device("mps")
else:
    device = torch.device("cpu")
print(f"Using device: {device}")

# ── Load preprocessed data ────────────────────────────────────
df = pd.read_csv("preprocessed_weather_data.csv")

# ── Features: exclude both targets and temperature targets ─────
TARGET_LOCATION = "location_name"
TARGET_COUNTRY  = "country"

EXCLUDE = [TARGET_LOCATION, TARGET_COUNTRY,
           "temperature_celsius", "feels_like_celsius"]

X = df.drop(columns=EXCLUDE).values
y_location = df[TARGET_LOCATION].values   # int labels
y_country  = df[TARGET_COUNTRY].values    # int labels

num_locations = len(np.unique(y_location))
num_countries = len(np.unique(y_country))
print(f"Unique locations: {num_locations} | Unique countries: {num_countries}")

# ── Train/test split & scaling ─────────────────────────────────
X_train, X_test, yl_train, yl_test, yc_train, yc_test = train_test_split(
    X, y_location, y_country,
    test_size=0.3, random_state=42, shuffle=True
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# ── Helper: build DataLoader ───────────────────────────────────
def make_loader(X_scaled, y, batch_size=64, shuffle=True):
    X_t = torch.tensor(X_scaled, dtype=torch.float32)
    y_t = torch.tensor(y,        dtype=torch.long)
    return DataLoader(TensorDataset(X_t, y_t),
                      batch_size=batch_size, shuffle=shuffle)

train_loc_loader = make_loader(X_train_scaled, yl_train, shuffle=True)
test_loc_loader  = make_loader(X_test_scaled,  yl_test,  shuffle=False)

train_cnt_loader = make_loader(X_train_scaled, yc_train, shuffle=True)
test_cnt_loader  = make_loader(X_test_scaled,  yc_test,  shuffle=False)

# ── Model ──────────────────────────────────────────────────────
class LocationClassifier(nn.Module):
    def __init__(self, input_dim, num_classes):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(128, 64),
            nn.ReLU(),

            nn.Linear(64, num_classes)   # raw logits → CrossEntropyLoss
        )

    def forward(self, x):
        return self.net(x)

input_dim = X_train_scaled.shape[1]

# ── Train/evaluate helper ──────────────────────────────────────
def train_and_evaluate(
    task_name, num_classes,
    train_loader, test_loader,
    epochs=30, lr=0.001
):
    model = LocationClassifier(input_dim, num_classes).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.5)

    with mlflow.start_run(run_name=f"Classify_{task_name}"):
        mlflow.log_params({
            "task": task_name,
            "epochs": epochs,
            "batch_size": 64,
            "learning_rate": lr,
            "num_classes": num_classes,
            "optimizer": "Adam",
            "loss": "CrossEntropyLoss"
        })

        # ── Training loop ──────────────────────────────────────
        for epoch in range(epochs):
            model.train()
            running_loss = 0.0

            for X_batch, y_batch in train_loader:
                X_batch, y_batch = X_batch.to(device), y_batch.to(device)

                optimizer.zero_grad()
                logits = model(X_batch)
                loss   = criterion(logits, y_batch)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            avg_loss = running_loss / len(train_loader)
            scheduler.step()
            mlflow.log_metric("train_loss", avg_loss, step=epoch)

            if (epoch + 1) % 5 == 0:
                print(f"[{task_name}] Epoch {epoch+1}/{epochs} | Loss: {avg_loss:.4f}")

        # ── Evaluation ─────────────────────────────────────────
        model.eval()
        all_preds, all_true = [], []

        with torch.no_grad():
            for X_batch, y_batch in test_loader:
                X_batch = X_batch.to(device)
                logits  = model(X_batch)
                preds   = torch.argmax(logits, dim=1).cpu().numpy()
                all_preds.extend(preds)
                all_true.extend(y_batch.numpy())

        acc = accuracy_score(all_true, all_preds)
        print(f"\n{'='*50}")
        print(f"[{task_name}] Test Accuracy: {acc:.4f} ({acc*100:.2f}%)")
        print(classification_report(all_true, all_preds, zero_division=0))
        print(f"{'='*50}\n")

        mlflow.log_metric("test_accuracy", acc)
        mlflow.pytorch.log_model(model, f"{task_name}_model")

    return model

# ── Run both classification tasks ──────────────────────────────
mlflow.set_experiment("Global_Weather_Location_Prediction")

print(">>> Training: location_name classifier")
model_location = train_and_evaluate(
    "location_name", num_locations,
    train_loc_loader, test_loc_loader,
    epochs=30, lr=0.001
)

print(">>> Training: country classifier")
model_country = train_and_evaluate(
    "country", num_countries,
    train_cnt_loader, test_cnt_loader,
    epochs=30, lr=0.001
)

# ── Inference on a single new sample ──────────────────────────
def predict_single(sample_row: dict, label_encoders: dict):
    """
    sample_row : dict of raw feature values (same columns as X)
    label_encoders : the dict of LabelEncoders from preprocessing
    Returns decoded location_name and country strings.
    """
    row_df   = pd.DataFrame([sample_row])
    row_scaled = scaler.transform(row_df.values)
    tensor   = torch.tensor(row_scaled, dtype=torch.float32).to(device)

    model_location.eval()
    model_country.eval()

    with torch.no_grad():
        loc_idx = torch.argmax(model_location(tensor), dim=1).item()
        cnt_idx = torch.argmax(model_country(tensor),  dim=1).item()

    location = label_encoders["location_name"].inverse_transform([loc_idx])[0]
    country  = label_encoders["country"].inverse_transform([cnt_idx])[0]

    print(f"Predicted location : {location}")
    print(f"Predicted country  : {country}")
    return location, country
```
